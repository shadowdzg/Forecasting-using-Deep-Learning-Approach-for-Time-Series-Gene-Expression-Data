```python
all_time = ["1-2", "1.5-2.5", "2-3", "2.5-3.5", "3-4", "3.5-4.5", "4-5", "4.5-5.5", "5-6", "5.5-6.5", "6-7", "7-8", "8-9", "9-10", "10-11", "11-12", "12-13", "13-14", "14-15", "15-16", "16-17", "17-18", "18-19", "19-20", "20-21", "21-22", "22-23", "23-24"]
exculded = ["1.5-2.5", "2.5-3.5", "3.5-4.5", "4.5-5.5", "5.5-6.5"]
time_point = [item for item in all_time if item not in exculded]
```

```python
df.rename(columns=dict(zip(df.columns, all_time)), inplace = True)
df
```

## Drop overlapped time points

```python
# df = df.drop(index=exculded)
df = df[time_point]
df
```

## Transpose matrix

#### in time series data the date must be `row index`

```python
df = df.T
df
```

```python
df = df.rename_axis(None, axis="columns")
df = df.rename_axis('time')
df
```

# Imputing NA

## get column that missing values

```python
gene_miss_tp = df.columns[df.isna().any()].tolist()
gene_miss_tp
```

```python
for col in gene_miss_tp:
    df[col].isna().sum()
    df[df[col].isna()][col]
```

## fill missing data

```python
for col in gene_miss_tp:
    df[col].interpolate(method="linear", inplace=True)
```
